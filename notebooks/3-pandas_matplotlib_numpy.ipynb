## NumPy

NumPy est une bibliothèque Python qui fournit des fonctionnalités de calcul numérique.
Elle est utilisée comme base pour d'autres bibliothèques Python comme Pandas, Matplotlib, etc.

Documentation officielle :
https://numpy.org/doc/stable/

NumPy est particulièrement utile pour les opérations sur les tableaux et les matrices.

```python
import numpy as np
```

```python
# Le array est un conteneur qui peut être initialisé
# avec une liste, une liste de listes, une liste de listes de listes, ...
# le niveau d'imbrication décrit le nombre de dimensions du array.
x = np.array([[1, 0.0, 3], [0, 1, 5]])
x
```

```python
# 'shape' informe sur la taille de chaque dimension
# Dans l'exemple, x contient 2 listes à 3 éléments.
x.shape
```

```python
# Contrairement aux conteneurs 'classiques', tous les éléments d'un array dovient être du même type.
# Dans l'exemple, des flottants.
x.dtype
```

```python
# Numpy dispose de types pour gérer des valeurs non-numériques spécifiques : "Not A Number", et "Infinity".
np.nan, np.inf
```

```python
# Ces types peuvent cohabiter avec des valeurs numériques
y = np.array([np.nan, 2], dtype=float)
print(y.dtype)
y
```

```python
# Des fonctions existent pour créer des array aux remplissages particuliers.
# Un array de 0
np.zeros((2, 3), dtype=int)
```

```python
np.ones((10, 4))
```

```python
arr = np.ones((10, 4)) * 15  # vectorisation => multiplication de chaque élément par 15 // PAS de boucle sur les éléments
arr
```

```python
np.empty((10, 4)) * np.nan
```

```python
np.arange(10.2, 1000.9, 99.9)
```

```python
np.empty_like(arr)  # ones_like(), zeros_like()
```

```python
np.identity(5)
```

```python
arr = np.array(np.arange(32), dtype='i1')
arr.shape
```

```python
# la forme peut être modifiée
arr2 =arr.reshape(2, 2, 2, 2, -1)
arr2.shape
```

```python
a = np.array([[1, 2, 3, 4, 5],[5, 8, 9, 2, 1]])
a
```

```python
# indexation
a[0]
```

```python
a[0][1]
# ou, mieux:
a[0, 1]
```

### base

```python
import matplotlib.pyplot as plt  # la formule magique
```

```python
# on se fabrique quelques données
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2 * np.pi * t)
```

```python
# on trace une courbe sommaire
plt.plot(t, s)
plt.show()  # cette ligne provoque l'affichage de la courbe
```

```python
# on peut évidemment décorer un peu tout ça
plt.plot(t, s)
plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('About as simple as it gets, folks')
plt.grid(True)
plt.show()
```

### interface pandas

```python
tdf = df.sort_index()
tdf.plot(y=["Température", "Humidité"])
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

tdf.plot(y="Température", ax=ax1, legend=False, linestyle="", marker=".", color="r")
tdf.plot(y="Humidité", ax=ax2, legend=False, linestyle="", marker=".", color="b")
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

tdf['2010':'2011'].plot(y="Température", ax=ax1, legend=False, linestyle="", marker=".", color="r")
tdf['2010':'2011'].plot(y="Humidité", ax=ax2, legend=False, linestyle="", marker=".", color="b")
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

rtdf = tdf[["Température", "Humidité"]].rolling("30d").median()

rtdf.plot(y="Température", ax=ax1, legend=False, color="r")
rtdf.plot(y="Humidité", ax=ax2, legend=False, color="b")
```
